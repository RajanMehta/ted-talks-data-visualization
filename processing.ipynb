```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('ted_main.csv', skipinitialspace = True)
```

```python
df.head(2)
```

```python
df['film_date'] = pd.to_datetime(df['film_date'],unit='s')
df['published_date'] = pd.to_datetime(df['published_date'],unit='s')
```

```python
df['duration'] = df['duration']/60.0
df['duration'] = df['duration'].round(decimals=2)
```

```python
df.head(2)
```

```python
df['name'].head()
```

```python
df['name'] = df['name'].str.split(': ', 1, expand=True)[1]
```

```python
df['name'].head()
```

```python
df['published_date'] = pd.to_datetime(df['published_date']).dt.date
```

```python
df.rename(columns={'duration': 'duration(mins)', 'name': 'title'}, inplace=True)
```

```python
df.head(2)
```

```python
df['ratings'] = df['ratings'].apply(lambda x: eval(str(x)))   #json string to json
```

```python
df['ratings'][1]
```

```python
maxRating = []
tedType = []

for i in range(len(df['ratings'])):
    c = max([d['count'] for d in df['ratings'][i]])
    maxRating.append(c)
    
    for j in range(len(df['ratings'][i])):
        if(df['ratings'][i][j]['count'] == maxRating[i]):
            tedType.append(df['ratings'][i][j]['name'])
            break
```

```python
len(tedType)
```

```python
df['popular_rating'] = tedType
df['popular_rating_count'] = maxRating
```

```python
df.head(2)
```

```python
df['related_talks'] = df['related_talks'].apply(lambda x: eval(str(x)))
df['related_talks'][0]
```

```python
related_titles = []

for i in range(len(df['related_talks'])):
    related_titles.append([d['title'] for d in df['related_talks'][i]])
```

```python
df['related_titles'] = related_titles
```

```python
df.to_csv('ted_updated.csv', encoding='utf-8')
```
